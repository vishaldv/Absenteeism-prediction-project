```python
import pandas as pd
import numpy as np
```

```python
pp_base=pd.read_csv('Absenteeism_project_preprocessed_file.csv')
```

```python
mldf=pp_base.copy()
mldf.head()
```

#### Classifying absenteeism into 'Excessively absent'(more than 3 hrs) and 'Moderately absent' (less than or equal to 3 hrs)

```python
#Finding median for Absenteeism in Hours
mldf['Absenteeism Time in Hours'].median()

#Take value '1' for employees absent for more than 3 hrs and '0' for less than or equal to 3 hrs
```

```python
targets=np.where(mldf['Absenteeism Time in Hours']>3,1,0)

# targets=<numpy_function>.where(if the values of 'Absenteeism in hours' is greater than 3,<store as 1>,<else store 0>)
```

```python
#Adding the new values ('0' & '1') to Absenteeism column
mldf['Excessive Absenteeism']=targets
mldf.head()
```

```python
targets.sum()/targets.shape[0]
#around 46% of the targets are 1's
```

```python
#Dropping 'Absenteeism in Hours'
data_with_targets=mldf.drop(['Absenteeism Time in Hours'],axis=1)
data_with_targets.head()
```

```python
#Selecting specific rows and columns
mldf.iloc[:6,:14]
#mlfd.iloc['row','column']
#mldf.iloc[':,0:14'] ':' means all rows, 0:14 means select columns 0 to 14
#mldf.iloc[4:8,5:10] '4:8' means rows 4 to 8, '5:10' means columns 5 to 10
```

```python
mldf.iloc[:6,:-1]
#. -1 means display all other columns except last column
#  -2  "           "          "        "     last 2 columns and so on 
```

```python
unscaled_inputs=data_with_targets.iloc[:,:-1]
unscaled_inputs.head()
```

### Standardising the data

```python
from sklearn.preprocessing import StandardScaler 
```

#### Checkpoint 1

```python
mldf1=unscaled_inputs.copy()
```

```python
mldf1.columns.values
```

```python
nodum=mldf1.copy()

nodum.head()
```

```python
cols_to_omit=['Reason1', 'Reason2', 'Reason3', 'Reason4', 'Education','Month value', 'Distance to Work' ]
```

```python
cols_to_scale=[x for x in nodum.columns.values if x not in cols_to_omit]
```

```python
absenteeism_scaler=StandardScaler(cols_to_scale)
#Creating StandardScaler object
```

```python
absenteeism_scaler.fit(nodum)
#this will calculate and store the mean & the standard deviation
#Standardisation info is contained in absenteeism_scaler
```

```python
#Applying the above scaling mechanism (transform)
scaled_inputs=absenteeism_scaler.transform(nodum)
# (.transform does the actual scaling)
```

```python
nodum.columns.values
```

#### Splitting the data into train, test and shuffle

```python
#Import module
from sklearn.model_selection import train_test_split
```

```python
#Split
train_test_split(scaled_inputs,targets)

#syntax 
# sklearn.mode_selection.train_test_split(inputs, targets, train_size*, shuffle=True**, random_state***)
# * train_size - value b/w 0 and 1 - denotes percentage of data to be trained
# ** shuffle - by default - True
# *** random_state - takes integer value - shuffles in random way
```

```python
#Assigning variables to above arrays
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets)
```

```python
#View shape
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

```python
#Assigning 80%-20% for training-testing
x_train,x_test,y_train,y_test=train_test_split(scaled_inputs,targets,train_size=0.8,random_state=20)
```

```python
print(x_train.shape, y_train.shape)
print(x_test.shape, y_test.shape)
```

#### Training the model and testing accuracy

```python
#import modules
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
```

```python
#Training the model
#Creating new variable which should be a logistic regression object
reg=LogisticRegression()
```

```python
#Fitting the regression
reg.fit(x_train,y_train)
```

```python
reg.score(x_train,y_train)
```

```python
#Manual check
model_outputs=reg.predict(x_train)
model_outputs
```

```python
feature_name=nodum.columns.values
```

```python
summary_table=pd.DataFrame(columns=['Feature name'],data=feature_name)
summary_table
```

```python
summary_table['Coefficient']=np.transpose(reg.coef_)
summary_table
```

```python
#Adding intercept on top of the df
summary_table.index=summary_table.index+1 #increases all indices by 1
```

```python
summary_table.loc[0]=['Intercept', reg.intercept_[0]]
```

```python
summary_table=summary_table.sort_index()
summary_table
```

```python
#Getting exponentials of all coefficients
summary_table['Odds ratio']=np.exp(summary_table.Coefficient)
summary_table
```

```python
summary_table.sort_values('Odds ratio',ascending=False)
```

A feature is not particular important:
- if co-efficient is around 0
- if odds ratio is around 1
